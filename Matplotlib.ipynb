```python
# !pip install matplotlib
#matplotlib--main module
# plt -- sub module
# import matplotlib.pyplot as plt

import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.array([0,2,20])
y=np.array([2,3,30])
plt.plot(x,y)
plt.show()
```

```python
z=np.array([10,1,23,4,55,6,878,67])
plt.plot(z,marker='o')
plt.show()
```

```python
z=np.array([10,1,23,4,55,6,878,67])
plt.plot(z,"s:r")
plt.show()
```

```python
z=np.array([10,1,23,4,55,6,878,67])
plt.plot(z,"*:b")
plt.show()
```

```python
z=np.array([10,1,23,4,55,6,878,67])
plt.plot(z,linestyle='dotted')
plt.show()
```

```python
z=np.array([10,1,23,4,55,6,878,67])
plt.plot(z,ls='-.')
plt.show()
```

```python
z=np.array([10,1,23,4,55,6,878,67])
plt.plot(z,linestyle='dotted',color='g')
plt.show()
```

```python
# increase line size

z=np.array([10,1,23,4,55,6,878,67])
x=np.array([4,24,56,8,98,12,76,34,98,25,64])
plt.plot(z,linewidth='20')
plt.plot(x,linewidth='10')
plt.show()
```

```python
# how to put lable and title

z=np.array([10,1,23,4,55,6,878,67])
x=np.array([4,24,56,8,98,12,76,34,98,25,64])
plt.plot(z,linewidth='20')
plt.plot(x,linewidth='10')
plt.title("z and x")
plt.grid()
plt.show()
```

```python
# build bar chart

a=np.array(['a','b','c','d','e'])
b=np.array([10,20,30,40,50])

plt.bar(a,b)
plt.show()
```

```python
a=np.array(['a','b','c','d','e'])
b=np.array([10,20,30,40,50])

plt.bar(a,b,color='red')
plt.show()
```

```python
a=np.array(['a','b','c','d','e'])
b=np.array([10,20,30,40,50])

plt.bar(a,b,color='yellow')
plt.show()
```

```python
# histogram

x=np.random.normal(120,60,30)
plt.hist(x)
plt.show()
```

```python

b=np.array([10,20,30,40,50])
mylabel=['blue','orange','apple','greenapple','grapes']
seperate=[0.2,0,0.2,0,0.2]

plt.pie(b,labels=mylabel,explode=seperate)
plt.show()
```

```python
# change colours

b=np.array([10,20,30,40,50])
mylabel=['blue','orange','apple','greenapple','grapes']
seperate=[0.2,0,0.2,0,0.2]
mycolors=['m','g','hotpink','k','b']

plt.pie(b,labels=mylabel,explode=seperate,colors=mycolors)
plt.show()
```
